```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
data=pd.read_csv("/kaggle/input/cancer/simulated_cancer_dataset.csv")
df=pd.DataFrame(data)
X =(df[["Age", "Tumor_Size_cm", "Radiation_Dose_Gy", "Gene_Expression_Level"]]).to_numpy()
Y=(df["Survival_Months"]).to_numpy()
```

```python
model=LinearRegression()
model.fit(X,Y)
```

```python
pr=model.predict(X)
```

```python
plt.scatter(Y,pr,label="True time", color="blue")
plt.plot([np.min(Y),np.max(Y)],[np.min(Y),np.max(Y)],label="Predicted time", color="red")
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("True vs Predicted months")
plt.legend()
plt.show()
```

```python
cost=mean_squared_error(Y,pr)
cost
```

```python
score = r2_score(Y, pr)
print("R2 Score:", score)
```
